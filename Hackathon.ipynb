```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
mapp =  {
    'UK':0,
    'SVK':1
}
df['Country of production'] = df['Country of production'].map(mapp)
```

```python
sns.heatmap(df.corr(), annot=True, cmap='Blues', fmt='.1g')
```

```python
from sklearn.model_selection import train_test_split
X= df.drop(columns=['Total Cost'])
y=df['Total Cost'] 
  
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,  shuffle=True) 
```

```python
test_X = pd.read_csv('test.csv')

test_X['Country of production'] = test_X['Country of production'].map(mapp)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from xgboost import XGBRegressor

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

models = {
    'Linear Regression': LinearRegression(),
    'Random Forest Regressor': RandomForestRegressor(),
    'Support Vector Regressor': SVR(),
    'K-Nearest Neighbors Regressor': KNeighborsRegressor(),
    'Decision Tree Regressor': DecisionTreeRegressor(),
    'AdaBoost Regressor': AdaBoostRegressor(),
    'Gradient Boosting Regressor': GradientBoostingRegressor(),
    'Lasso Regression': Lasso(),
    'Ridge Regression': Ridge(),
    'Elastic Net Regression': ElasticNet(),
    'XGBoost Regressor': XGBRegressor()
}
mae_score = 10000000000
best_model = ''

for model_name, model in models.items():
    try:
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)
        mae = mean_absolute_error(predictions, y_test)
        if mae< mae_score:
            mae = mae_score
            best_model = model
        print(f"Mean Absolute Error for {model_name}: {mae}")
        y_pred = model.predict(test_X)
        print(f'Predictions for Test Data {y_pred}')
        print('\n')
    except:
        pass
```

```python

import joblib 
  
  
joblib.dump(best_model, 'price_prediction.pkl') 
  
knn_from_joblib = joblib.load('price_prediction.pkl') 
  
knn_from_joblib.predict(X_test) 
```
